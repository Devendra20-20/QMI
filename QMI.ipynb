## Introduction

The branch of mechanics which plays a fundamental role in the description
and understanding of natural phenomena related to very small (atomic or
subatomic) scale is called Quantum mechanics.
for eg: the existance and the properties of atoms, the chemical bond and
the propagation of an electron in a crystal will be expalined by quantum
mechanics.
The basis of our present understanding of all natural phenomena, including
those traditionally treated in chemistry, biology,etc is called Quantum
mechanics.
Most of the atomic and molecular phenomena are well explained by the
non-relativistic quantum mechanics.

## One particle wave function space
From the probabilistic interpretation we infer that
$\int d^3r |\psi(r,t)|^2 = 1$
where the integration extends over all space. We are lead to study a
class of square-integrable functions. This class of functions
constitutes $L^2$ space and has a structure of Hilbert space. This space
is too wide in scope and we will be interested in only those functions,
which are continuous and infinitely differentiable for the simple reason
that a function, discontinuous at some point, is inconsistent with the
probabilistic interpretation. The functions we shall consider also may
have bounded domain. This set of functions, which are solutions of
Schrodinger equation, is denoted by $\phi$ and is composed of suciently
regular functions of $L^2$. Clearly, $\phi$ is a subset of $L^2$.
